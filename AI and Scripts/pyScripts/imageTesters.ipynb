```python
from PIL import Image
import matplotlib.pyplot as plt
from torchvision import datasets, transforms
from torchvision.transforms import v2
```

## Testing the CenterCrop

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

image_path = 'Data/train/Soja - Masaico Amarelo/YM (110).jpg'
image = Image.open(image_path)

center_crop = transforms.CenterCrop(size=(900, 900))
cropped_image = center_crop(image)

ax[0].imshow(image)
ax[0].set_title("Imagem Original")
ax[0].axis('off')

ax[1].imshow(cropped_image)
ax[1].set_title("Imagem Recortada")
ax[1].axis('off')

plt.show()
```

## Testing the RandomResizedCrop

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

image_path = 'Data/train/Soja - Masaico Amarelo/YM (110).jpg'
image = Image.open(image_path)

# Sobre o Resize: scale=(Tamanho Mnimo, Tamanho Mximo)
randomResizedImage = transforms.RandomResizedCrop(128, scale=(0.1, 1.0))
resizedImage = randomResizedImage(image)

ax[0].imshow(image)
ax[0].set_title("Imagem Original")
ax[0].axis('off')

ax[1].imshow(resizedImage)
ax[1].set_title("Imagem Recortada")
ax[1].axis('off')

plt.show()
```

## Testing the Resize

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

image_path = 'Data/train/Soja - Masaico Amarelo/YM (110).jpg'
image = Image.open(image_path)

# Sobre o Resize: scale=(Tamanho Mnimo, Tamanho Mximo)
FixResize = transforms.Resize((128, 128))
resizedImage_Fix = FixResize(image)

ax[0].imshow(image)
ax[0].set_title("Imagem Original")

ax[1].imshow(resizedImage_Fix)
ax[1].set_title("Imagem Recortada")

plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

image_path = 'Data/train/Soja - Masaico Amarelo/YM (110).jpg'
image = Image.open(image_path)

grayscale_transform = v2.Grayscale()
changedImage = grayscale_transform(image)

ax[0].imshow(image)
ax[0].set_title("Imagem Original")
ax[0].axis('off')

ax[1].imshow(changedImage, cmap='gray')
ax[1].set_title("Imagem em Tons de Cinza")
ax[1].axis('off')

plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

image_path = 'Data/train/Soja - Masaico Amarelo/YM (110).jpg'
image = Image.open(image_path)

transform = transforms.Compose([
    transforms.Grayscale(num_output_channels=1),  # Convert to grayscale
    transforms.Resize((64, 64)),                 # Resize images to 64x64
    transforms.ToTensor(),                       # Convert to tensor
    transforms.Normalize((0.5,), (0.5,))         # Normalize to [-1, 1]
])

changedImage = transform(image)

# Denormalize and convert to NumPy array
denormalized_image = changedImage * 0.5 + 0.5  # Reverse normalization
denormalized_image = denormalized_image.squeeze(0).numpy()  # Remove channel dimension

ax[0].imshow(image)
ax[0].set_title("Imagem Original")
ax[0].axis('off')

ax[1].imshow(denormalized_image, cmap='gray')
ax[1].set_title("Imagem Alterada")
ax[1].axis('off')

plt.show()
```
